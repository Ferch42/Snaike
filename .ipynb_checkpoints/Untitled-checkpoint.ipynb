```python
import pygame,sys
from pygame.locals import *
```

```python
#colors
WHITE = (255, 255, 255)
BLACK=(0,0,0)
```

```python
def desloc(x,y,ori):
    des=1
    if(ori=='r'):
        return (x+des,y)
    elif(ori=='l'):
        return (x-des,y)
    elif(ori=='u'):
        return (x,y-des)
    elif (ori=='d'):
        return (x,y+des)
```

```python
#body
class body:
    def __init__(self,x,y,ori):
        self.posix=x
        self.posiy=y
        self.orientation=ori
        self.turns=[]
    
    def move(self):
        
        if(len(self.turns)==0):
            self.posix,self.posiy=desloc(self.posix,self.posiy,self.orientation)
            
        else:
            turnx,turny,turno= self.turns[0]
            virtualx,virtualy=desloc(self.posix,self.posiy,self.orientation)
            
            if(((self.orientation=='r' and virtualx>turnx) or (self.orientation=='l' and virtualx<turnx) 
               or (self.orientation=='u' and virtualy<turny) or (self.orientation=='d' and virtualy>turny))
              and self.orientation!=turno):
                
                self.posix,self.posiy=desloc(self.posix,self.posiy,turno)
                self.orientation=turno
                self.turns.pop(0)
                
            else:
                self.posix,self.posiy=desloc(self.posix,self.posiy,self.orientation)
                
```

```python
#Snake
class Snake:
    
    def __init__(self,x,y,ori='r'):
        self.headx=x
        self.heady=y
        self.blocks=[]
        self.orientation=ori   
        self.headImg=pygame.Surface((0,0))
        self.bodyImg=pygame.Surface((0,0))
        
    def move(self,ori):
        
        if(ori!=self.orientation):
            #turn happened
            turn=(self.headx,self.heady,ori)
            for block in self.blocks:
                block.turns.append(turn)
            self.orientation=ori
   
        self.headx,self.heady=desloc(self.headx,self.heady,ori)
        
        for b in self.blocks:
            b.move()
    
    def calculateSnakeArea(self):
    
        area=[]
        b=self.bodyImg
    
        for bloc in self.blocks:
            u=bloc.posix-int(b.get_width()/2)
            d=bloc.posix+int(b.get_width()/2)
            r=bloc.posiy+int(b.get_height()/2)
            l=bloc.posiy-int(b.get_height()/2)
            area.append([u,d,r,l])
        return area
    
    
    def isAlive(self):
        
        area=self.calculateSnakeArea()
        if(self.orientation=='r' or self.orientation=='l'):
            hx=self.headx
            hyu=self.heady-int(self.bodyImg.get_height()/2)
            hyd=self.heady+int(self.bodyImg.get_height()/2)
            for co in area:
                if(hyd<=co[0] or hyu>=co[1]):
                    continue
                if(hx==co[3] or hx==co[2]):
                    return False
        
        elif(self.orientation=='u' or self.orientation=='d'):
            hxr=self.headx+int(self.bodyImg.get_width()/2)
            hxl=self.headx-int(self.bodyImg.get_width()/2)
            hy=self.heady
            for co in area:
                if(hxl>=co[2] or hxr<=co[3]):
                    continue
                if(hy==co[0] or hy==co[1]):
                    return False
        return True
```

```python
def calcVirtualPosition(lastx,lasty,ori,i,headImg,bodyImg):
    
    orientation=ori
    virtualx,virtualy=0,0
    
    if(orientation=='r'):
        virtualx, virtualy = (lastx-10-bodyImg.get_width()),lasty
            
    elif(orientation=='l'):
        if(i!=0):
            virtualx,virtualy = (lastx+bodyImg.get_width()+10),lasty
        else:
            virtualx,virtualy = (lastx+headImg.get_width()+10),lasty
        
    elif(orientation=='u'):
        virtualx,virtualy= lastx, (lasty-10 - bodyImg.get_height())
        
    elif(orientation=='d'):
        if(i!=0):
            virtualx,virtualy = lastx,(lasty +10+ bodyImg.get_height())
        else:
            virtualx,virtualy = lastx,(lasty +10+ headImg.get_height())
    
    return (virtualx,virtualy)
```

```python
def renderSnake(display,snake):
    
    headImg=snake.headImg
    bodyImg=snake.bodyImg
    
    #add head
    display.blit(headImg,((snake.headx-int(headImg.get_width()/2)),(snake.heady-int(headImg.get_height()/2))))
    
    #add body
    for body in snake.blocks:
        display.blit(bodyImg,((body.posix-int(bodyImg.get_width()/2)),(body.posiy-int(bodyImg.get_height()/2))))
      
    pygame.display.update()
        
        
```

```python
#init
pygame.init()
dim=600
displaysurf=pygame.display.set_mode((dim,dim))
pygame.display.set_caption('Snaike')
fpsClock = pygame.time.Clock()
FPS = 120

#emilia variables
emiliaImg = pygame.image.load('blue.png')
emiliaSnake = Snake(int(dim/2),int(dim/2))
emiliaSnake.headImg=emiliaImg

#subaro
subaroImg=pygame.image.load('blue.png')
emiliaSnake.bodyImg=subaroImg
emiliaSnake.blocks.append(body(emiliaSnake.headx-10-subaroImg.get_width(),emiliaSnake.heady,emiliaSnake.orientation))
emiliaSnake.blocks.append(body(emiliaSnake.headx-10-subaroImg.get_width()-10-subaroImg.get_width(),emiliaSnake.heady,emiliaSnake.orientation))
t=10+subaroImg.get_width()
emiliaSnake.blocks.append(body((emiliaSnake.headx-t-t-t),emiliaSnake.heady,emiliaSnake.orientation))
emiliaSnake.blocks.append(body((emiliaSnake.headx-t-t-t-t),emiliaSnake.heady,emiliaSnake.orientation))
emiliaSnake.blocks.append(body((emiliaSnake.headx-t-t-t-t-t),emiliaSnake.heady,emiliaSnake.orientation))
emiliaSnake.blocks.append(body((emiliaSnake.headx-t-t-t-t-t-t),emiliaSnake.heady,emiliaSnake.orientation))
orientation='r'

#game loop
while True:   
    displaysurf.fill(WHITE)
    
    for event in pygame.event.get():
        if event.type== QUIT:
            pygame.quit()
            sys.exit()
        elif event.type==KEYDOWN:
            if event.key==K_UP and orientation!='d':
                orientation='u'
            elif event.key==K_DOWN and orientation!='u':
                orientation='d'
            elif event.key==K_RIGHT and orientation!='l':
                orientation='r'
            elif event.key==K_LEFT and orientation!='r':
                orientation='l'
        
    #emilia motion
    emiliaSnake.move(orientation)
    if(not emiliaSnake.isAlive()):
        print("game over")
    
    renderSnake(displaysurf,emiliaSnake)
    fpsClock.tick(FPS)

            
```
